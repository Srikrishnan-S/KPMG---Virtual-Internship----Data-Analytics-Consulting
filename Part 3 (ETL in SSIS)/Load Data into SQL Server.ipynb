```python
connection_str = f'mssql+pyodbc://{user}:{password}@{server}/{database}?driver=ODBC+Driver+17+for+SQL+Server'

try:
    engine = create_engine(connection_str)
    
    Transactions_Table.to_sql(name = 'Transactions', con = engine, if_exists = 'replace', index = False)

    CustomerDemographic_Table.to_sql(name = 'CustomerDemographic', con = engine, if_exists = 'replace', index = False)
    
    CustomerAddress_Table.to_sql(name = 'CustomerAddress', con = engine, if_exists = 'replace', index = False)

    NewCustomer_Table.to_sql(name = 'NewCustomerList', con = engine, if_exists = 'replace', index = False)
    
    print("Data export to SQL Server database is successful")
except Exception as e:
    print("Error connecting to SQL Server database: ", str(e))
```
