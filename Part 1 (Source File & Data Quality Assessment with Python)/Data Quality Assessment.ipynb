**DUPLICATE VALUES :**

```python
# Assessing Duplicate Records:
class ProcessDuplicates:
    def __init__(self, df):
        self.df = df
        
    def duplicate_count(self):
        return self.df.duplicated().sum()
    
    def duplicate_rate(self):
        rate = self.df.duplicated().sum() * 100 / len(self.df)
        return rate
    
    def duplicated_records(self):
        duplicates = self.df[self.df.duplicated(keep = False)]
        return duplicates
    
    def remove_duplicates(self):
        return self.df.drop_duplicates(inplace = True)
    
```

**NULL VALUES :**

```python
# Assesssing Null Values: 
class ProcessNulls:
    def __init__(self, df):
        self.df = df

    def null_count(self):
        return self.df.isnull().sum()

    def null_rate(self):
        rate = self.df.isnull().sum() * 100 / len(self.df)
        return rate

    def detect_null_features(self):
        if_null_df = pd.DataFrame({
            "Column_Name": [feature for feature in self.df.columns], 
            "If_Null": [self.df[feature].isnull().any() for feature in self.df.columns]
        })
        return if_null_df

    def percentage_null(self):
        null_percentage = pd.DataFrame({
            "Column Name": [feature for feature in self.df.columns], 
            "Percentage Null": [self.df[feature].isnull().mean() * 100 for feature in self.df.columns]
        })
        return null_percentage

    def plot_null(self, vertical=True):
        plot_df = self.percentage_null()
        if vertical:
            plot = sns.barplot(x="Column Name", y="Percentage Null", data=plot_df)
        else:
            plot = sns.barplot(y="Column Name", x="Percentage Null", data=plot_df)
        return plot
    
    def null_records(self):
        null_records = self.df[self.df.isnull().any(axis = 1)]
        return null_records

    def remove_nulls(self):
        return self.df.dropna(inplace=True)

    def impute_null(self, features, method='mean', value=None):
        for feature in features:
            if method == 'mean':
                self.df[feature].fillna(self.df[feature].mean(), inplace=True)
            elif method == 'median':
                self.df[feature].fillna(self.df[feature].median(), inplace=True)
            elif method == 'mode':
                self.df[feature].fillna(self.df[feature].mode().iloc[0], inplace=True)
            elif method == 'custom' and value is not None:
                self.df[feature].fillna(value, inplace=True)
```
